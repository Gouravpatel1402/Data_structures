# Linked List

```python
class linked_list:
  def __init__(self):#empty linked list
    self.head=None
    self.n=0 #number of nodes

  def __len__(self):
       return self.n

  def insert_head(self,value):
      node=Node(value)#new node
      node.next=self.head #creating connection
      self.head= node #reassigning head
      self.n+=1 #updating length

  def traverse(self):
      current=self.head
      result=''
      while(current!=None):
          result=result+str(current.data)+"-->"
          current=current.next
      return result[:-3]

  def insert_tail(self,value):
      new_node=Node(value)
      if self.head==None:
         #for an empty linked list
         self.head=new_node
         self.n+=1
         return

      curr= self.head #initial pointer
      # reaching to the tail:
      while curr.next!=None:
        curr=curr.next
      curr.next=new_node#we were at the tail
      self.n+=1
  def insert_after (self,key,value):
      new_node=Node(value)
      current=self.head
      while current.next!=None:
        if current.data==key:
          break
        current=current.next
      new_node.next=current.next
      current.next=new_node
      self.n+=1

  def clear(self):
      self.head=None
      self.n=0

  def delete_head(self):
      self.head=self.head.next
      self.n-=1

  def pop(self):
      if self.n==1:
        self.head=None
      curr=self.head
      while curr.next.next!=None: #reaching second last element condition
        curr=curr.next
      curr.next=None

  def remove(self,value):
      if self.head.data==value:
        self.head=self.head.neaxt
      curr= self.head
      while curr.next.data!=value:
        curr=curr.next
      curr.next=curr.next.next
      self.n-=1
  def find(self,value):
      index=0
      current=self.head
      while current.data!=value:
        current=current.next
        index+=1
      return index
  def __getitem__(self,index): #indexing
      curr=self.head
      for i in range(index):
        curr=curr.next
      return curr.data
```
